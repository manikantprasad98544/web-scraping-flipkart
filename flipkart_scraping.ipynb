```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
link='https://www.flipkart.com/search?q=tv&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_8_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_8_0_na_na_na&as-pos=8&as-type=TRENDING&suggestionId=tv&requestId=9c9fa553-b7e5-454b-a65b-bbb7a9c74a29'
```

```python
page = requests.get(link)
print(page)
```

```python
content=page.content
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
title=soup.find_all('div',{'class':'_4rR01T'})
```

```python
product=[]
for titles in title:
    product.append(titles.text)
print(product)
len(product)
```

```python
topics=soup.find_all('div',{"class":"_2kHMtA"})
```

```python
len(topics)
```

```python
product=[]
price=[]
ratings=[]
apps=[]
link=[]
os=[]
refresh_rate=[]
speaker=[]

for data in topics:
    #geting product name
    name=data.find('div',{'class':'_4rR01T'})
    product.append(name.text)
    
    #getting product price
    prices=data.find('div',{'class':'_30jeq3 _1_WHN1'})
    price.append(prices.text)
    
    #getting product rating
    rating=data.find('div',{'class':'_3LWZlK'})
    ratings.append(rating.text)
    
    #getiing product link
    base_url="https://www.flipkart.com"
    url=data.find('a',{'class':'_1fQZEK'})
    link.append(base_url+url['href'])
    
    #apps,os,speaker,refersh rate suported by product
    speci = data.find('div', attrs={'class':'fMghEO'})
    for each in speci:
        data1=each.find_all('li', attrs={'class':'rgWa7D'})
        app =data1[0].text
        os_data = data1[1].text
        speaker_data=data1[3].text
        refresh_rate_data=data1[4].text
    apps.append(app)
    os.append(os_data)
    speaker.append(speaker_data)
    refresh_rate.append(refresh_rate_data)
```

```python
df=pd.DataFrame({'Product Name':product,'Link':link,'Price':price,'Supported Apps':apps,'OS':os,'Refresh Rate':refresh_rate,'Speaker':speaker,'Rating':ratings})
```
